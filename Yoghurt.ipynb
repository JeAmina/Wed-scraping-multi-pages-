```python
reqvesty
```

```python
tekst = reqvesty.text
```

```python
BS = BeautifulSoup(tekst, "html.parser")
```

```python
ssylki = BS.find_all(class_='col-xs-offset-1')
```

```python
listt = []
for item in ssylki:
    vse=item.find_all('a')
    for tag_a in vse:
        link = 'https://contentservice.kz' + tag_a.get('href')
        listt.append(link)
listt
```

```python
listt = listt[0:108]
```

```python
data = []
```

```python
inform = []
product_inf = []
```

```python
for catalog in listt:
    request_link = requests.get(catalog)
    content = BeautifulSoup(request_link.text, 'html.parser')
    pagination_is_Done = False

    urls = [catalog]
    for stranicy in urls:
 
        rlink = requests.get(stranicy)
        content_stranicy = BeautifulSoup(rlink.text, 'html.parser')
        block = content_stranicy.find('div', class_='catalog__grid-view')
        aaa = block.find_all('a', class_='link-gray')

        product_urls = []
        for a_tag in aaa:
            product_urls.append('https://contentservice.kz' + a_tag['href'])


        for product_url in product_urls:
            
            rl = requests.get(product_url)
            product_main_page = BeautifulSoup(rl.text, 'html.parser')

            tags = product_main_page.find('div', class_='row-eq-height')
            tags = tags.find_all('div', class_='col-lg-9')

            title = tags[0].find('h1').get_text(strip=True)

            types = product_main_page.find('ol', class_='breadcrumb')
            types = types.find_all('a')

            tags = tags[0].find_all('a')
            category = tags[-2].get_text(strip=True)
            subcategory = tags[-1].get_text(strip=True)

            fabric = product_main_page.find('ul', class_='list-lg-pad')  
            try:
                izgotovitel = fabric.find('a').get_text(strip=True)
            except:
                izgotovitel = ''
            try:
                country = fabric.find('span').get('title')
            except:
                country = ''
            print(category,title, subcategory, izgotovitel, country, product_url)
            product_inf.append([category,title, subcategory, izgotovitel, country, product_url])
           
```

```python
df = pd.DataFrame(product_inf, columns=['Nazvanie','Category','Type','Izgotovitel','Country','url_of_product'])
df.to_csv('HW4.csv', index=False,encoding='utf-8')
```
